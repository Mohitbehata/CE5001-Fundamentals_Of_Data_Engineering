 We have to create a CSV file with randomly generated data and save it as 'sample_data.csv'.

```python
from faker import Faker
import csv

# Select the locale for fake data
fake = Faker('en_US')

# Open a CSV file to write (UTF-8 encoding to support emojis)
with open('sample_data.csv', mode='w', encoding='utf-8') as sample_data:
    writer = csv.writer(sample_data)

    # Write the header row
    writer.writerow(['Name', 'State', 'Plate', 'SSN', 'Occupation', 'Emoji'])

    # Generate and write 1000 rows of random data
    for _ in range(1000):
        writer.writerow([fake.name(), fake.state_abbr(), fake.license_plate(), fake.ssn(), fake.job(), fake.emoji()])
```

- Create a Faker object named "fake" and specify the locale as 'en_US' to generate data in English with US convention.
- Open a CSV file named "sample_data.csv" for writing, using UTF-8 encoding to support emojis.
- Create a CSV writer object called "writer" to write data to the CSV file.
- Write the header row to the CSV file, specifying the column names. using **fake.__()** (1000 rows of randomly generated fake data, with columns for name, state abbreviation, license plate number, SSN, occupation, and emoji.)


```python
import pandas as pd
df = pd.read_csv('/content/sample_data.csv')
df.head()
```

- Using pandas to handle the csv file as a dataframe for further procedure.

```python
df.drop(df[df['State'] != 'MO'].index, inplace = True)
df.head()
```

- From the dataframe we have we are removing all such rows where state is not MO.

```python
def formatting_ssn(x):
    return x.replace('-', '')
```

- We are creating a function to formatting_ssn as per the task given.
- We are removing all '-' and combining the ssn into single number from "xxx-xxx-xxxx"to "xxxxxxxxx".



```python
df['SSN'] = df['SSN'].apply(lambda x: formatting_ssn(x))
df.head()
```

- The lambda function takes an input x and applies the formatting_ssn function to it.
- The result of applying the function is assigned back to the 'SSN' column in the DataFrame, effectively replacing the original SSNs with the formatted versions.

```python
df.to_csv('output_data.tsv', sep="\t")
```

```python
df.to_json('output_data.json')
```
