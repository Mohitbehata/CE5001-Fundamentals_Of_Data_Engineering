Problem 2 [10pts] The following code cell prints unicode emoji of Euro currency symbol. Print the all currency symbols included in https://en.wikipedia.org/wiki/Currency_Symbols_(Unicode_block).

```python
import unicodedata

def print_currency_symbols():
    # Range of Unicode code points for currency symbols
    start_code_point = 0x20A0
    end_code_point = 0x20CF

    for code_point in range(start_code_point, end_code_point + 1):
        try:
            # Get the character for the code point
            char = chr(code_point)

            # Check if it's a currency symbol
            if unicodedata.category(char) == 'Sc':
                print(f"Code Point: U+{code_point:04X}, Symbol: {char}")

        except ValueError:
            # Some code points might not be valid characters
            pass

if __name__ == "__main__":
    print_currency_symbols()
```

Problem 6 [10pts] Query the customer.db to find how many customers you have in each country. Your result set should have two columns including Country and Count. Sort your result set with Count column in descending order.

```python
import pandas as pd
import sqlite3
from tabulate import tabulate

# Connect to the existing SQLite database "customer.db".
conn = sqlite3.connect("customer.db")

# Query: Count customers in each country and sort the result by count in descending order.
query = """
    SELECT Country, COUNT(*) AS Count
    FROM customers
    GROUP BY Country
    ORDER BY Count DESC;
"""

# Execute the query and fetch the result into a Pandas DataFrame.
result = pd.read_sql_query(query, conn)

# Display the result using tabulate.
print("Number of customers in each country (sorted by count in descending order):")
print(tabulate(result, headers='keys', tablefmt='psql'))
```

```python
%%writefile mapper.py
#!/usr/bin/env python3

import sys

# Read input from standard input
for line in sys.stdin:
    # Split the line into words
    words = line.strip().split()

    # Output each word with a count of 1
    for word in words:
        print(f"{word}\t1")

```

```python
%%writefile reducer.py
#!/usr/bin/env python3

import sys

current_word = None
current_count = 0

# Read input from standard input
for line in sys.stdin:
    # Split the input into word and count
    word, count = line.strip().split('\t', 1)

    # Convert count to integer
    count = int(count)

    # If the current word is the same as the previous word, increment count
    if word == current_word:
        current_count += count
    else:
        # If the current word is different, output the result for the previous word
        if current_word:
            print(f"{current_word}\t{current_count}")
        # Set current word and count to new values
        current_word = word
        current_count = count

# Output the result for the last word
if current_word:
    print(f"{current_word}\t{current_count}")
```

Problem 9 [15pts] Using pyspark, count the total number of words given in https://raw.githubusercontent.com/nivdul/spark-in-practice-scala/master/data/wordcount.txt

```python
!pip install pyspark
import requests
from pyspark.sql import SparkSession
from pyspark.sql.functions import split, explode, col

# Initialize Spark session
spark = SparkSession.builder.appName("TotalWordCount").getOrCreate()

# URL of the document
url = "https://raw.githubusercontent.com/nivdul/spark-in-practice-scala/master/data/wordcount.txt"

# Download the file
response = requests.get(url)
content = response.text

# Split the content into lines

lines = content.strip().split()

# Create DataFrame directly from the list of strings
data = spark.createDataFrame(lines, "string").toDF("value")

# Split each line into words and explode the array of words into separate rows
words_df = data.select(explode(split(data.value, " ")).alias("word"))

# Filter out any empty words
words_df = words_df.filter(col("word") != "")

# Count the total number of words
total_word_count = words_df.selectExpr("count(*) as total_word_count").collect()[0]["total_word_count"]

# Show the result
print(f"Total Word Count: {total_word_count}")
```
