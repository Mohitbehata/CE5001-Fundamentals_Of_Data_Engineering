####**Printable Characters in Ascii and their Hex values**

```python
for i in range(0x21,0x7f):
  i_hex = hex(i)
  bytes_val = i.to_bytes(1, 'big')
  print(f'Hex: {i_hex} Char: {bytes_val.decode("ascii")}')
```

- The iteration through all ascii printable characters in hexadecimal is performed for i in range(0x21, 0x7f).

- Hexadecimal representation of integer i is stored in i_hex using i_hex = hex(i).

- In bytes_val, the integer i is converted to a single-byte value in big-endian byte order.

- The print function prints the hexadecimal value of i as well as the ascii encoded character (bytes_val.decode("ascii")).

- A hexadecimal string is iterated through to find i in range(0x21, 0x7f):: Iterates through the range of ascii printable characters.

- Hexadecimal representation of integer i is stored in i_hex using i_hex = hex(i).

- A single-byte value is converted from the integer i into bytes_val by doing i.to_bytes(1, 'big').

- In this print function, the hexadecimal value of i is printed, and the character decoded as ascii is printed as well.

####**Printable Characters in cp1252 and their Hex values**

```python
for i in range(0x21,0xff):
  try:
    i_hex = hex(i)
    bytes_val = i.to_bytes(1, 'big')
    print(f'Hex: {i_hex} Char: {bytes_val.decode("cp1252")}')
  except UnicodeDecodeError:
    pass
```

- The iteration through all cp1252 printable characters in hexadecimal is performed for i in range(0x21, 0xff). We have extended the range for cp1252 because it has more printable characters.

- Hexadecimal representation of integer i is stored in i_hex using i_hex = hex(i).

- In bytes_val, the integer i is converted to a single-byte value in big-endian byte order.

- The print function prints the hexadecimal value of i as well as the cp1252 encoded character (bytes_val.decode("cp1252")).

- A hexadecimal string is iterated through to find i in range(0x21, 0xff):: Iterates through the range of cp1252 printable characters.

- Hexadecimal representation of integer i is stored in i_hex using i_hex = hex(i).

- A single-byte value is converted from the integer i into bytes_val by doing i.to_bytes(1, 'big').

- In this print function, the hexadecimal value of i is printed, and the character decoded as cp1252 is printed as well.

-  Uses a try...except block to catch and ignore UnicodeDecodeError exceptions for bytes that cannot be decoded.
